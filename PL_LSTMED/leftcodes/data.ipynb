```python
import pandas as pd
import numpy as np
import matplotlib as plt
import pandas_datareader.data as web


def boll_bands(data1,data2, ndays):
    """
    计算布林带
    :param data: 股票的df格式数据
    :param ndays: 计算使用的简单移动均线周期
    :return:
    """
    ma = pd.Series(np.round((data1['close']-data2['close']).rolling(ndays).mean(), 2), name='MA%s'%ndays)  # 计算nday均线
    # pandas.std() 默认是除以n-1 的，即是无偏的，如果想和numpy.std() 一样有偏，需要加上参数ddof=0
    # 此处添加ddof的原因是wind和yahoo的计算均采用的有偏值进行的计算
    std = pd.Series(np.round((data1['close']-data2['close']).rolling(ndays).std(ddof=0), 2))  # 计算nday标准差，有偏
    b1 = ma + (2 * std)  # 此处的2就是Standard Deviations
    B1 = pd.Series(b1, name='UpperBollingerBand')
    data1 = data1.join(ma)  # 上边不写name 这里报错
    data1 = data1.join(B1)

    b2 = ma - (2 * std)
    B2 = pd.Series(b2, name='LowerBollingerBand')
    data1 = data1.join(B2)
    MA = pd.Series(ma, name='SMA(20)')
    data1 = data1.join(MA)

    return data1
```

```python
data1 = pd.read_csv("沪深300.csv", encoding='ISO-8859-1')
data2 = pd.read_csv("futures_main_sina_hist.csv", encoding='ISO-8859-1')

data2.head
```

```python

n = 20
NIFTY_BBANDS = boll_bands(data1,data2, n)
print(NIFTY_BBANDS.loc[:, ['LowerBollingerBand', 'UpperBollingerBand']])
print(NIFTY_BBANDS.columns)

x = NIFTY_BBANDS['date']
```

```python
NIFTY_BBANDS['date'] = pd.to_datetime(NIFTY_BBANDS['date'],format='%Y/%m/%d')
NIFTY_BBANDS.index = NIFTY_BBANDS['date'] #转为index
NIFTY_BBANDS = NIFTY_BBANDS.drop('date', axis='columns')#删除重复的date列
NIFTY_BBANDS.head()
#df.shape
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set()
%matplotlib inline
warnings.filterwarnings('ignore')
plt.figure(figsize=(8,6),dpi=80,num=4)
myfont=matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\simsun.ttc')
x = NIFTY_BBANDS.index
y_low = NIFTY_BBANDS['LowerBollingerBand']
y_high = NIFTY_BBANDS['UpperBollingerBand']
y_diff = data1['close']-data2['close']
y_sma = NIFTY_BBANDS['SMA(20)']
plt.plot(x, y_low, color='r', label = 'LowerBollingerBand');
plt.plot(x, y_high, color='g', label = 'UpperBollingerBand');
plt.plot(x, y_diff, color='b', label = 'price gap');
plt.plot(x, y_sma, color='g', label = 'SMA(20)');
plt.xlabel('日期',FontProperties=myfont,fontsize=15)
plt.ylabel('价格差',FontProperties=myfont,fontsize=15)
plt.title('沪深300指数-IF股指期货价差布林带',FontProperties=myfont,fontsize=15)
plt.legend()
plt.grid()
plt.savefig("沪深300.png", transparent=True)
plt.show()
```
